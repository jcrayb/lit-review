```python
import pandas as pd

c = pd.read_csv('current.csv')
a = pd.read_csv('all_papers.csv')

combined = pd.DataFrame(columns = c.columns)

for r in a.iloc:
    name = r.Name

    if r.Name in c.Name.to_list():
        crow = c.loc[c.Name ==name]
        crow.Hyperlink = r.Hyperlink

    else:
        val = r.to_list()

        val[-2] = ""

        crow = pd.DataFrame(columns = c.columns, index = [0])
        crow.loc[crow.index == 0] = val


        #print(crow)
    if len(str(crow.Authors).split(', '))>2:
        crow.Authors = (str(crow.Authors).split(', ')[0] + " et al.")[2:]
    combined = pd.concat([combined, crow])

    combined.to_csv('combined.csv')
```
